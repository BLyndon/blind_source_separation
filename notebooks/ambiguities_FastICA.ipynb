```
from sklearn.decomposition import FastICA
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```
def bss(sources):
  N = len(sources.T)

  S_true = sources

  W = np.random.rand(N,N)
  X = np.dot(S,W.T)

  ica = FastICA(n_components=N)
  S_ = ica.fit_transform(X)
  W_ = ica.mixing_

  labels = ["MIXED", "TRUE", "ICA"]
  signals = [X.T, S_true.T, S_.T]

  plt.figure(figsize=(10, 7), dpi= 80)
  for i, (label, sigs) in enumerate(zip(labels, signals), 1):
    plt.subplot(len(labels), 1, i)
    plt.ylabel(label)
    for sig in sigs:
      plt.plot(sig)
  plt.show()
```

```
n_samples=10000
t_0=0
t_1=15

time = np.linspace(t_0, t_1, n_samples)
s1 = np.sin(3*time)
s2 = np.sign(np.sin(8*time))
s3 = signal.sawtooth(5*time)

S = np.c_[s1, s2, s3]
```

The change of amplitude corresponds to the scaling ambiguity and the change of color corresponds to permutation ambiguity.

```
bss(S)
```
